```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pylab import rcParams
rcParams['figure.figsize'] = 12,9
```

```python
player_data = pd.read_csv('player_data.csv')
player_data['year_fight']=player_data['year_end'] - player_data['year_start']
players = pd.read_csv('Players.csv')
players.dropna(inplace=True)
players.drop('Unnamed: 0', axis=1, inplace=True)
players['born'] = players.born.astype('int64')
```

```python
player_data.head()
```

```python
players.head()
```

# График №1 
Топ-10 городов рождения играков

```python
city = players.groupby('birth_city')[['Player']].count().sort_values(by = 'Player', ascending=False).head(10)
```

```python
city.plot(kind='bar', rot=45)
```

```python
city.style.bar()
```

# График №2
Распределение кол-во играков по годам рождения

```python
born_yers = players.groupby('born')[['Player']].count().sort_values(by = 'born')
```

```python
born_yers.plot()
```

# График №3
Топ-5 Университетов

```python
university = players.groupby('collage')[['Player']].count().sort_values(by = 'Player', ascending=False).head(5)
university.plot(kind='pie', y='Player')
```

```python
university.style.bar()
```
